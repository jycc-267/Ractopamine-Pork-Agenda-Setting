```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
from tomorrow import threads
from datetime import datetime
import time
import re
import random
import pandas as pd
```

```python
ua = UserAgent()
headers = {'User-Agent': ua.random}
# 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'

@threads(10)
def Multithreads(url):
    try:
        return requests.get(url, headers=headers)
    except:
        pass
```

```python
# https://www.ftvnews.com.tw/search/%e8%90%8a%e8%b1%ac/195

links = []
head = 'https://www.ftvnews.com.tw'

for i in range(1, 195, 1):
    url = 'https://www.ftvnews.com.tw/search/%e8%90%8a%e8%b1%ac/{}'.format(i)
    try:
        rawdata = Multithreads(url)
        time.sleep(random.uniform(1, 5))
    except:
        pass
    soup = BeautifulSoup(rawdata.text, 'html.parser')
    for i in soup.select('.news-list ul li'):
        link = i.find('a').get('href')
        links.append(head+link)
```

```python
url = 'https://www.ftvnews.com.tw/news/detail/2021B15P10M1'
rawdata = Multithreads(url)
soup = BeautifulSoup(rawdata.text, 'html.parser')
```

```python
date = datetime.strptime(soup.find('meta', attrs={'property':'article:published_time'})['content'], '%Y-%m-%dT%H:%M:%S+08:00')
title = re.sub('\n|\r|\t|\xa0', '', soup.find('h1').text.strip())
content = ""
target = soup.find('div', {'class':'article-body'}).find_all('p')[:-3]
for p in target:
    content += re.sub('\n|\r|\t|\xa0|\u3000', '', p.getText().strip())
print(title, date, content)
```

```python
all_posts = []

for link in links:
    try:
        response = Multithreads(link)
        time.sleep(random.uniform(5, 10))
        soup = BeautifulSoup(response.text, 'html.parser')
        title = re.sub('\n|\r|\t|\xa0', '', soup.find('h1').text.strip())
        date = datetime.strptime(soup.find('meta', attrs={'property':'article:published_time'})['content'], '%Y-%m-%dT%H:%M:%S+08:00')
        content = ""
        target = soup.find('div', {'class':'article-body'}).find_all('p')[:-3]
        for p in target:
            content += re.sub('\n|\r|\t|\xa0|\u3000', '', p.getText().strip())
        
        all_posts.append({'title':title, 
                          'date':date, 
                          'link':link, 
                          'content':content})
    except:
        pass
```

```python
ftv_politics = pd.DataFrame(all_posts)
ftv_politics = ftv_politics.drop_duplicates(subset = ['link']).reset_index(drop = True)
# ftv_politics = ftv_politics.sort_values('date', ascending = False).reset_index(drop = True)
ftv_politics
```
